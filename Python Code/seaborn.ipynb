```python
import seaborn as sns
```

```python
# basic plotting 
tips = sns.load_dataset('tips')
tips
```

```python
## scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
plt.show()
```

```python
sns.lineplot(x = 'sex', y = 'tip', data = tips, marker = '*', color = 'teal')
plt.show()
```

```python
sns.barplot(x = 'time', y = 'total_bill', data = tips)
plt.show()
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
plt.show()
```

```python
# violine plot
sns.violinplot(x = 'day', y = 'total_bill', data = tips)
plt.show()
```

```python
#histogram
sns.histplot(tips['total_bill'], bins = 10, kde = True) #kde - curve
```

```python
 ## kde plot
 sns.kdeplot(tips['total_bill'], fill = True)
```

```python
## Pair plot
# relationship between to variable with different variables 
sns.pairplot(tips)
```

```python
## co-relation with heat map
corr = tips[['total_bill', 'tip', 'size']].corr()
corr
```

```python
sns.heatmap(corr, annot = True, )
```
