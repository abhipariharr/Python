# Pandas

Pandas is an open-source software library built on python for data analysis and data manipulation.
Pandas library provides data structure designed specifically to handle tabular dataset with simplified python API.
- To preocess and manipulate tabular data, implementing operation such as a loading, aligning, merging, and transforming data set.

Pandas is a python library used to:
- work with tables of data.
- Read CSV/Excel files.
- Clean, analyze and manipulate data easily.

# Series

```python
# Series is a similar to a column in a table.

import pandas as pd

data = [10,4,7,3]
series = pd.Series(data)
print("Series\n", series)
print(type(series))
```

```python
#dictionary series

dict = {'a': 1, 'b' : 2, 'c' : 3}
series_dict = pd.Series(dict)
print(series_dict)
```
