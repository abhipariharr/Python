- To preocess and manipulate tabular data, implementing operation such as a loading, aligning, merging, and transforming data set.

Pandas is a python library used to:

- Clean, analyze and manipulate data easily.

# Series

```python
# Series is a similar to a column in a table.

import pandas as pd

data = [10,4,7,3]
series = pd.Series(data)
print("Series\n", series)
print(type(series))
```

```python
#dictionary series

dict = {'a': 1, 'b' : 2, 'c' : 3}
series_dict = pd.Series(dict)
print(series_dict)
```

```python
# DataFrame
## creating a dataframe from a dictionary of list

data = {
    'Name' : ['Abhishek', 'Ansh', 'Priyanshu'],
    'Age' : [22, 23, 21],
    'City' : ['Aligarh', 'Delhi', 'Ghaziabad']
}
df = pd.DataFrame(data)
print(df)
print(type(df))
```
