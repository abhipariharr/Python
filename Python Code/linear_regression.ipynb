```python
#Importing Libraries
import numpy as np
from sklearn.linear_model import LinearRegression

#Preaparing Data
x = np.array([1,2,4,6]).reshape(-1, 1)   #Study hours
y = np.array([0,0,1,1]) # fail = 0, pass = 1

#Creating model
model1 = LinearRegression()

#Train Model
model1.fit(x, y)

#Predicting
prediction = model1.predict([[5]])
print("Prediction: ", prediction)
```
