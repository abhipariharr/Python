```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4] # x-axis
y = [1,4,5,9] # y-axis

## creating aline plot
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Baic Line Grapgh')
```

```python
x = [1,2,3,4] # x-axis
y = [1,4,5,9] # y-axis

## creating aline plot
plt.plot(x,y, color = 'Red', linestyle = '-')
```

```python
x = [1,2,3,4] # x-axis
y = [1,4,5,9] # y-axis

## creating aline plot
plt.plot(x,y, color = 'Red', linestyle = '--', marker = '*', linewidth = 3, markersize = 10)
```

```python
## creating multiple plots

x = [1,2,3,4]
y1 = [1,3,5,8]
y2 = [7,5,6,9]

plt.figure(figsize=(9,5))

## defining rows and columns
plt.subplot(1,2,1)
plt.plot(x,y1, color = 'green', marker = 'o')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Plot 1')

## 2nd column graph
plt.subplot(1,2,2) #last 2 is for plot
plt.plot(x,y2, color = 'red', marker = '*')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Plot 2')
```

```python
# Creating 4 graphs

## creating multiple plots

x = [1,2,3,4]
y1 = [1,3,5,8]
y2 = [7,5,6,9]

plt.figure(figsize=(9,5))

## defining rows and columns
plt.subplot(2,2,1)
plt.plot(x,y1, color = 'green', marker = 'o')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Plot 1')

## 2nd column graph
plt.subplot(2,2,2) #last 2 is for plot
plt.plot(x,y2, color = 'red', marker = '*')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Plot 2')

## 3rd column graph
plt.subplot(2,2,3) #last 2 is for plot
plt.plot(y1,y2, color = 'blue', marker = 'o')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Plot 3')

## 4th column graph
plt.subplot(2,2,4) #last 2 is for plot
plt.plot(y2,y1, color = 'Yellow', marker = '*')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Plot 4')
```

## Bar Chart

```python
year = [1990, 2000, 2010, 2020]
sales = [30, 54, 65, 87]

plt.bar(year, sales, color = 'blue')
plt.xlabel('Year')
plt.ylabel('Sales Percentage')
plt.title('Sales Analysis')
```
